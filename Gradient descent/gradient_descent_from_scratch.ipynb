# Gradient Descent from Scratch

In this notebook, I implement linear regression using gradient descent.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)
```

## 1. Data Generation
We will generate synthetic data for simple linear regression: $y = 4 + 3x + noise$

```python
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Visualize the data
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Synthetic Data')
plt.show()
```

## 2. Initialization
Initialize separate parameters for w (weight/slope) and b (bias/intercept).

```python
# Initialize w and b randomly
w = np.random.randn(1)
b = np.random.randn(1)

print(f"Initial w: {w}")
print(f"Initial b: {b}")
```

## 3. Training Loop
We will implement the gradient descent steps inside the loop:
- Prediction: $\hat{y} = Xw + b$
- Loss (MSE): $J(w,b) = \frac{1}{n} \sum (\hat{y} - y)^2$
- Gradients:
    - $\frac{\partial J}{\partial w} = \frac{2}{n} \sum (\hat{y} - y)x$
    - $\frac{\partial J}{\partial b} = \frac{2}{n} \sum (\hat{y} - y)$

```python
# Hyperparameters
learning_rate = 0.1
n_epochs = 1000

losses = []

for epoch in range(n_epochs):
    # 1. Compute predictions
    y_pred = X.dot(w) + b
    
    # 2. Compute MSE loss
    n = len(X)
    loss = (1/n) * np.sum((y_pred - y)**2)
    losses.append(loss)
    
    # 3. Compute gradients
    dw = (2/n) * np.sum((y_pred - y) * X)
    db = (2/n) * np.sum(y_pred - y)
    
    # 4. Update parameters
    w = w - learning_rate * dw
    b = b - learning_rate * db
    
    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}")

print(f"\nFinal w: {w}")
print(f"Final b: {b}")
```

## 4. Visualization
Let's look at the learning curve and the final fit.

```python
# Plot Learning Curve
plt.plot(losses)
plt.xlabel("Epochs")
plt.ylabel("MSE Loss")
plt.title("Gradient Descent Training Loss")
plt.show()

# Plot Final Fit
plt.scatter(X, y, label='Data')
plt.plot(X, X.dot(w) + b, color='red', label='Prediction')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title("Linear Regression Fit")
plt.show()
```
